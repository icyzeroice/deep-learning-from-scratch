# Use library `numpy` to generate array

````python
import numpy as np


print('\r\n----- generate -----')
a = [1.0, 2.0, 3.0]
x = np.array(a)

print(x)
print(type(x))


print('\r\n----- multiply -----')
b = [2, 4, 6]
y = np.array(b)

z = x * y
print(z)


print('\r\n----- 2 dimensional array (matrix) -----')
A = [[1, 2], [3, 4]]
B = [[4, 3], [2, 1]]

X = np.array(A)
Y = np.array(B)

print(A, B, '\r\n', X, '\r\n', Y, X.shape, Y.shape)
print(X.dtype, Y.dtype)
print(X + Y)
print(X * Y)

print('\r\n----- 广播 -----')
print(10 * X)
print(X * np.array([1, 2]))
print(np.array([1, 2]) * X)
````

# Use `matplotlib` to draw charts

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 6, 0.1)

y = np.sin(x)
plt.plot(x, y, label = "sin")

y2 = np.cos(x, y)
plt.plot(x, y2, linestyle = "--", label = "cos")

plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")

# 标注说明
plt.legend()

plt.show()
````

# Operate Matrix

````python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

B = np.array([
    [2, 3, 4],
    [4, 5, 6]
])

print(A.shape, B.shape)

C = np.dot(A, B)

D = np.dot(B, A)

print(C)
print(D)


XA = np.array([
    [1, 2, 3],
    [2, 3, 4]
])

XB = np.array([
    [1, 2],
    [2, 3],
    [4, 5],
    [6, 7]
])
````
